### 1. Computational Thinking

- Before digging into Python, it's important to realize that computational thinking is language-agnostic.
- It has to do more with logic and abstraction and less with specific syntax. However, a language with good syntax (like Python) can make it incredibly easier to understand and implement the logic of a computational problem we want to solve. 
- A "program" is essentially a collection of instructions to a computer.
- Programming is basically talking computationally via a series of instructions, which we can do in many languages.
- Talking computationally requires thinking computationally.
- Thinking computationally: the goal is to make the computer do as much of our work as possible, as in the fewest commands possible.
- Think about having a list of numbers and having to add 1 to each. It’s ok if our list is \[1,2,3,4,5\]. We can just produce \[1+1, 2+1, 3+1, 4+1, 5+1\] (even this is annoying).

- But imagine a million numbers. Or solving a complex model with thousands of parameters and conditions. Or solving a model like this for each of the 5 million users of your app.

- We must to figure out the best (smarter, fastest, most elegant, or some other criteria) to do these things by giving orders to our computer.

- This is what computational programming is about, and computational thinking is a skill you build with practice, patience, and understanding, that consistis of gaining an intuition of how to best solve computational problems by logically decomposing the bigger problems into smaller ones.
- Correctly formulating the problem (in your head or in paper) is the first and maybe the most important step.

### 2. If-else statements

- Probably the most important programming expression, not onkly in Python but in any language.

- "if-else" statements have the following syntax (including the ":" and the indentation):

            if <logical condition>:

- In Python's logic, the "else" part is not strictly necessary.

### 3. For loops

- Loops provide a mechanism for repeating work in a program.
- Loops are often used when we need to process a set of values and perform the same action for each.
- For example, given a list of prices, we may want to compute the tax on each price as well as keep a running total of all the taxes paid.
- Of course, we don't want to do this manually with a calculator, for each price in a long list! That's what programs are for.
- So we will use a for loop.
- "for loops" have the following syntax:
            for <variable> in <sequence>:
                <statements>

#### Example 1: Print each price in the list

```python
for p in [10, 25, 5, 70, 10]:
    print("The price is", p)
```

Of course, for loops can contain any kind of statments or expression, including if statements:

#### Example 2: Check the age

```python
mylist = [10, 25, 5, 70, 10]
age_limit = 18

for age in mylist:
    if age >= age_limit: 
        print("The age", age, "is high enough")
    else:
        print("The age", age, "is not high enough")
```

#### Example 3:  Price discrimiantion: checking the age of a customer before letting them in to the restaurant and then calculate the fee according to age:

- A more complicated example that will make use of many things we have learned until now:

```python
ages = [10, 25, 5, 70, 10] # our list of prices
total_fees_collected = 0 # starting value before we accumulate total tax by executing the for loop

for age in ages:
    print("Checking age...", age) # we will check p
    if age >= 18:
        print("Old enough. Calculate and then charge entry fee")
        fee = 0.10 * age # so we calculate the fee depending on the person's age
        total_fees_collected = total_fees_collected + fee # acummulate the total fees 
                             # This variable's value will no longer be zero once it gets to this point
    else:
        print("Not old enough. No entry. No fee collected")
        print("---------")
    print("The total fees collected is", total_fees_collected)
```

```python
total_fees_collected # we check the most recent value of our collected fees variable:
```

#### Example 4: Check if a number is odd or even:

```python
for n in range(1, 11): # range is a useful built-in Python function that will give us the sequence of integers between the first and second elements
    if (n % 2) == 1:
        print(n, "es non")
    else:
        print(n, "es par")
```

#### The break statement: 

- What if we want to interrupt and quit the loop as soon as something happens (this is, when a logical condition is met)? This could be to save some unnecessary computing time por because we just want to find the first occurrence of something (such as an error) and then interrupt the program.

- In Python we can use the break statement for this. Example:


```python
mylist = [22, 3, 11, 67, 12, 45, 500000, 67, 799, 1, 2]

for i in sorted(list(range(0, len(mylist))), reverse=True):
    if mylist[i] > 1000:
        print("LARGE VALUE FOUND!:", mylist[i], ". NO NEED TO KEEP LOOKING, LET'S BREAK THE LOOP.")
        break
    else:
        print("LET'S KEEP LOOKING FOR A LARGE NUMBER - ", mylist[i], "IS NOT LARGE ENOUGH!")
```

```python
myvec = (22, 3, 11, 67, 12, 45, 500000, 67, 799, 1, 2)
type(myvec)
for num in myvec:
    if num > 1000:
        print("LARGE VALUE FOUND!:", num, ". NO NEED TO KEEP LOOKING, LET'S BREAK THE LOOP.")
        break
    else:
        print("LET'S KEEP LOOKING FOR A LARGE NUMBER - ", num, "IS NOT LARGE ENOUGH!")
```

### 2. Dictionaries
- Another great Python data structure.
- Weird name (the reference to the lexical dictionary makes some sense, but much more generally a Python dictionary is a __correspondence__ between a unique key and a given value we define for it.
- More specifically, a set of key: value pairs, with the requirement that the keys are unique (within one dictionary).
- Just as \[\]  creates an empty list, {} creates an empty dictionary, which we can then fill in with comma-separated  key:value pairs within the braces.
- General sytantic structure: 

            {(unique)key: value}
            
- A dict lets us look quickly and intuitively for values via their associated keys.
- This may sound simple, but it lets you do a lot of things super efficiently and logically! Example: 
    - Imagine you have a dataset of countries and several variables, and you need to add an additional column contiening the capital the country.
    - You can add it manually. But what if the dataset has 5 million rows (every country appears many times).
    - You can have a short dictionary like {country: capital} and then create your column as a function of the country column using this dict.

#### Defining a function

#### Example 1: Exponentiate a number

- Following the syntax above, let's define a function that squares number 2:

```python
def square_num2(): # no specific input is also valid: think about this as a tranformation from "None" to square of 2
    return 2**2
```

Now let's make a call to this function:

```python
square_num2()
```

Not the most useful function ever. Why? We can generalize more:

```python
def square_num(x): # let x be any number we want!
    return x**2
square_num(90)
```

```python
square_num(5)
```

Remember lists and for loops?

```python
sq_list = []
for num in range(1,50):
    sq_list = sq_list + [square_num(num)]
```

```python
sq_list
```

Much better. We can generalize more!

```python
def elevate_num(x, p): # let x be any number we want, and also power any number we want
    return x**p
elevate_num(3, 1) # we must respect the order of args, OR make explicit which is which (elevate_num(p=4, x=3))
```

```python
Much better! Still, we must be careful of possible unexpected results when defining functions:
```

```python
elevate_num(-3, 0.5)
```

- Notice we just tried to get the square root of a negative number, which gives us complex numbers as a results.

- This is mathematically correct, but maybe we don't want this type of result in our particular case of use - what if we are doing arithmetic operations only applicable to real numbers with the output of our function and then the program crashes?
- We can then easily control what happens with if-else statements, and give our function a more complex structure:

```python
def elevate_num(x, p):
    if x < 0 and p < 1:# let x be any number we want, and also power any number we want
        print("INPUTS NOT VALID", x, p)
        return None
    else:
        return x**p
elevate_num(-3, 0.5)
```

The fact that we return __None__ is very important - it covers the program from errors. We use print just to get a "notification" that an expected "error" happened:

```python
power_list = []
for x in range(-10,50):
    for p in [0.5, 2, 3]:
        power_list = sq_list + [elevate_num(x, p)]
```

```python
power_list
```

#### Example 2:  Total collected fees by restaurants

- Back to our fees-collecting task, and to dictionaries, we can create a nice function:

```python
def calc_total_fee(ages, age_limit=18, age_tax =0.10):
    total_fees_collected = 0 # starting at zero
    for age in ages:
        if age >= age_limit:
            fee = age_tax * age 
            total_fees_collected = total_fees_collected + fee # acummulate the total fees 
                                 
            
    return round(total_fees_collected)
```

```python
ages_r1_g1 = [1, 2, 5, 70, 10]
ages_r2_g1 = [67, 88, 35, 20, 19, 13, 56]

ages_r1_g2 = [10, 25, 35, 13, 45, 46]
ages_r2_g2 = [20, 25, 5, 70, 10, 22, 48, 23, 24, 25]

# Two groups of restaurants, each owns 2 restaurants who collect their fees
rests_group1_dict = {"rest1_group1": ages_r1_g1, "rest2_group1" : ages_r2_g1}
rests_group2_dict = {"rest1_group2": ages_r1_g2, "rest2_group2" : ages_r2_g2}
```

```python
rests_group1_dict["total_collected"] = calc_total_fee(rests_group1_dict["rest1_group1"]) + calc_total_fee(rests_group1_dict["rest2_group1"]) 
rests_group2_dict["total_collected"] = calc_total_fee(rests_group2_dict["rest1_group2"]) + calc_total_fee(rests_group2_dict["rest2_group2"])
```
